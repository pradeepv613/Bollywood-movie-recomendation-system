```python
import imdb

moviesDB = imdb.IMDb()

# # Help?
# print(dir(moviesDB))
# ----------------------------------------
# 1) Search for a title
movies = moviesDB.search_movie('inception')

# print('Searching for "inception":')
# for movie in movies:
#     title = movie['title']
#     year = movie['year']
#     print(f'{title} - {year}')

## Help?
#print(movies[0].keys())
    
```

```python
def get_genre(x):
  genres = []
  
  try:
    movies = moviesDB.search_movie(x)
    id = movies[0].getID()
    movie = moviesDB.get_movie(id) 
    genres = movie['genres'] 
    direcStr= ' '.join(map(str,genres))
    genres.append(direcStr)
    return genres
  except:
    print("An exception occurred")
```

```python
df32['Genre'] = df32['Title'].map(lambda x: get_genre(str(x)))
```
