```python
import pandas as pd
import numpy as np
import requests
import bs4 as bs
import urllib.request

link = "https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2019"
source = urllib.request.urlopen(link).read()

soup = bs.BeautifulSoup(source,'lxml')



tables = soup.find_all('table',class_='wikitable')

len(tables)
```

```python

df2 = pd.read_html(str(tables[2]))[0]
df3 = pd.read_html(str(tables[3]))[0]
df4 = pd.read_html(str(tables[4]).replace("'1\"\'",'"1"'))[0] # avoided "ValueError: invalid literal for int() with base 10: '1"'
df5 = pd.read_html(str(tables[5]).replace("'1\"\'",'"1"'))[0] 
```

```python
df = df2.append(df3.append(df4.append(df5,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
#getting genre for 2020 movies using imdb API
def get_genre(x):
  genres = []
  movies = moviesDB.search_movie(x)
  id = movies[0].getID()
  movie = moviesDB.get_movie(id) 
  try:
      genres = movie['genres'] 
      direcStr= ' '.join(map(str,genres))
      genres.append(direcStr)
      return genres
  except:
    print("An exception occurred")
```

```python
df['Genre'] = df['Title'].map(lambda x: get_genre(str(x)))
```
